##Class Node

````python
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None
````

###Creating Nodes with Class Node

````python
n1 = Node(6)
n2 = Node(4)
n3 = Node(5)
n4 = Node(7)
n5 = Node(87)
````

###Linking Nodes

````python
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
````

##Class Linked List

````python
class LinkedList:
  def __init__(self,head):
    self.head = head

  #Print LinkedList (Recursion):
  def PrintLL(self,head):
    if head == None:
      print('X')
      return
    else:

      print(head.data,end="->")
      self.PrintLL(head.next)

  #Sum of Nodes (Recursion):
  def Sum_NodesLL(self,head):
    if head == None:
      return 0
    else:
      return head.data + self.Sum_NodesLL(head.next)

  #Length of LL (Recursion):
  def LengthLL(self,head):
    if head == None:
      return 0
    else:
      return 1 + self.LengthLL(head.next)

  #Insert at end
  def Insert_EndLL(self,head,value):
    if head == None:
      n = Node(value)
      self.head=n
      return

    elif head.next == None:
      n = Node(value)
      head.next = n
      return

    else:
      self.Insert_EndLL(head.next,value)

  #Insert at end (Alternate Method)
  def Insert_endAlternate(self, head, val):
    if head is None:
      return Node(val)
    else:
      head.next = self.Insert_endAlternate(head.next, val)
      return head
````

````python
def swap_nodes_in_pair(head):
  if head == None or head.next==None:
    return head
  else:
    temp = head
    head = head.next
    temp.next = head.next
    head.next = temp

    head.next.next = swap_nodes_in_pair(head.next.next)

    return head
````
